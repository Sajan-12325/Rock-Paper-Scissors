```python
## Rules:
# Rock beats Scissors
# Scissors beat Paper
# Paper beats Rock

import random
Game = ['Rock', 'Paper', 'Scissors']

Computer_score = 0
User_score = 0

def start_game(Computer, User):
    global Computer_score, User_score
    
    if Computer == User:
        print("It's a tie, Let's continue to next round")   
    elif (Computer == 'Rock' and User == "Scissors") or (Computer == 'Scissors' and User == 'Paper') or (Computer == 'Paper' and User == 'Rock'):
        Computer_score += 1 
        print("Computer Wins this round!\n")
    else:
        User_score += 1
        print("You Win this round!\n")
number = 5
while number >0:
    
        Computer = random.choice(Game)   
        User = input('Users choice:  ').title()
        if User not in Game:
            print("Invalid choice! Please enter Rock, Paper, or Scissors.\n")
            continue  # Restart loop if input is invalid

        print("Computer's Choice: ", Computer)

        winner = start_game(Computer, User)
    
        number -= 1
        print(f"Score -> You: {User_score} | Computer: {Computer_score}\n")
            
print("\nFinal Score:")
print(f"User: {User_score} | Computer: {Computer_score}")
            
if User_score > Computer_score:
    
    print("Congratulations! You won the game!")
elif User_score < Computer_score:
    print("Computer won the Game!!!!!, All the best for your next game.")
else:
    print("It's a draw, You both played really well.")
        
```
